```python
import pandas as pd
# Re-loading to ensure we are working with the correct dataframe
data = pd.read_csv('/content/student_placement_dataset (2).csv')
data['Certifications']= data['Certifications'].map({'No': 0, 'Yes': 1})
data.head()
```

```python
data = pd.read_csv('/content/student_placement_dataset (2).csv')
data['Certifications'] = data['Certifications'].map({'No': 0, 'Yes': 1})
data['Placed'] = data['Placed'].map({'No': 0, 'Yes': 1})
# Drop rows with NaN in Placed if any still exist
data.dropna(subset=['Placed'], inplace=True)
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data = pd.get_dummies(data, columns=['Communication_Skills', 'Coding_Skills'], drop_first=True, dtype=int)
data.head()
```

```python
data.shape
```

```python
data.drop(['Student_ID'], axis=1, inplace=True)
data.head()
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split

# The 'data' DataFrame is already preprocessed: Student_ID dropped,
# Certifications and Placed mapped, and Communication_Skills, Coding_Skills dummified.
# So, we can directly use the 'data' DataFrame to create x and y.

y = data['Placed']
x = data.drop(['Placed'], axis=1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
scaler = StandardScaler()
x_train_scled = scaler.fit_transform(x_train)
x_test_scled = scaler.transform(x_test)

logistic_model = LogisticRegression()
logistic_model.fit(x_train_scled, y_train)
y_pred = logistic_model.predict(x_test_scled)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(x_train_scled, y_train)
y_pred = logistic_model.predict(x_test_scled)
y_pred
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np

metric_k=[]
neighbors=np.arange(3,15)

for k in neighbors:
  knn_classifier=KNeighborsClassifier(n_neighbors=k,metric='minkowski')
  knn_classifier.fit(x_train,y_train)
  y_pred=knn_classifier.predict(x_test)
  accuracy=accuracy_score(y_test,y_pred)
  metric_k.append(accuracy)

metric_k
```

```python
import matplotlib.pyplot as plt
plt.plot(neighbors,metric_k,'o-')
plt.xlabel('K value')
plt.ylabel('accuracy')
```

```python
knn_classifier=KNeighborsClassifier(n_neighbors=9,metric='minkowski')
knn_classifier.fit(x_train,y_train)
y_pred=knn_classifier.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
metric_k.append(accuracy)
print(accuracy)
```

```python
confusion_matrix(y_test,y_pred)
```
